```python
###Esta parte del codigo se demora 32 minutos 
import os
from PyPDF2 import PdfReader

# La ruta de la carpeta que contiene los archivos PDF
pdf_folder = "/content/drive/MyDrive/Proyecto integrador/in"

# Leer todos los archivos PDF en la carpeta
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

# Extraer el texto de cada archivo PDF
pdf_texts = []
titles = []
for pdf_file in pdf_files:
    with open(os.path.join(pdf_folder, pdf_file), 'rb') as f:
        reader = PdfReader(f)
        pdf_text = ' '.join(page.extract_text() for page in reader.pages)
        pdf_texts.append(pdf_text)
        titles.append(pdf_file[:-4])  # Eliminar la extensión ".pdf"

print(f"{len(pdf_texts)} archivos PDF leídos.")
```

# **Bloque *3*: Tokenización y limpieza de textos**

Tokenización de corpus , es decir, dividir los diferentes textos en palabras individuales.


```python
##tiempo 3 min
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import os
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

def clean_text(text):
    # Eliminar las contracciones
    # text = re.sub(r"n't", " not", text)
    # text = re.sub(r"'s", " is", text)
    # Tokenizar el texto
    tokens = nltk.word_tokenize(text)
    # Eliminar las palabras vacías, los espacios y las url
    stop_words = set(stopwords.words("english"))
    tokens = [token for token in tokens if token.isalpha() and token.lower() not in stop_words and not re.match(r"https?://\S+", token)]
    # Eliminar los guiones y las palabras muy cortas o muy largas
    tokens = [token.replace("-", "") for token in tokens if 2 < len(token) < 18]

    #eliminar como palabras exactas
    words_to_remove = ["also", "jpmorgan", "obtai", "ocally", "rbh", "unfav" , "pmpsa", "emmanuel", "gyy"]
    tokens = [token for token in tokens if token.lower() not in words_to_remove]

    # Aplicar el stemming
    # stemmer = PorterStemmer()
    # tokens = [stemmer.stem(token) for token in tokens]
    return tokens
# Usar la función para limpiar cada texto en la lista 'pdf_texts'
cleaned_texts = [clean_text(text) for text in pdf_texts]
```

# **Bloque 4: Métodos de vectorización Aplicar : TF-IDF, BOW , Gensim** **texto en negrita**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convertir la lista de listas en una lista de strings
stemmed_texts_joined = [' '.join(text) for text in cleaned_texts]

# Crear una instancia de TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Transformar los datos procesados en una matriz TF-IDF
X_tfidf = tfidf_vectorizer.fit_transform(stemmed_texts_joined)

print(f"Matriz TF-IDF creada con dimensiones {X_tfidf.shape}")
```

# **Bloque 6: Estandarización de conjunto de datos**

**Estandarizar los datos**

En K-Means significa llevar a cabo un proceso de ajuste de la escala de las características (variables) de un conjunto de datos antes de aplicar el algoritmo K-Means. La estandarización es una técnica de preprocesamiento de datos que busca transformar las variables en una escala común, generalmente con **media 0 y desviación estándar 1.**

La estandarización es importante en K-Means porque este algoritmo se basa en la distancia entre los puntos de datos para determinar la asignación a los grupos (clusters). Si las características tienen diferentes escalas o unidades de medida, puede haber un sesgo en la asignación de puntos a los grupos, ya que las características con escalas más grandes tendrán un mayor impacto en el cálculo de las distancias.

**tener en cuenta que al usar el parámetro with_mean=False estás evitando que se reste la media a cada valor, lo cual puede ser necesario si quieres aplicar algún algoritmo que asuma que los datos están centrados. Además, debes asegurarte de que el conjunto de datos no sea disperso (sparse), ya que la estandarización puede alterar la estructura de los datos y hacerlos más densos.**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(X_tfidf)
```

```python
import numpy as np
#Verificar si el conjunto de datos es disperso
print(f"El conjunto de datos es disperso: {X_tfidf.nnz / np.prod(X_tfidf.shape) < 0.1}")

#Verificar si la estandarización altera la estructura de los datos
print(f"La estandarización altera la estructura de los datos: {X_scaled.nnz / np.prod(X_scaled.shape) != X_tfidf.nnz / np.prod(X_tfidf.shape)}")
```

El resultado indica que el conjunto de datos es disperso, es decir, que tiene muchos valores cero. Esto puede ser conveniente para ahorrar espacio de memoria y tiempo de cómputo. También indica que la estandarización no altera la estructura de los datos, es decir, que no cambia el número de valores cero o no cero. Esto puede ser bueno para mantener la consistencia de los datos. Sin embargo, debes tener en cuenta que la estandarización no centra los datos, lo cual puede ser un problema si quieres aplicar algún algoritmo que asuma que los datos tienen una media cero. # **Bloque 7: Exploración de datos (distribucion normal?)**

**La divergencia de Kullback-Leibler (KL)**

Es una medida que nos permite comparar qué tan diferentes son dos distribuciones de probabilidad. Es decir, nos ayuda a entender cuánta información se pierde si tratamos de representar una distribución de probabilidad (P) usando otra distribución (Q).

**Algunas características importantes de la divergencia KL son:**

No negatividad: La divergencia KL siempre es mayor o igual a cero. Esto significa que si las dos distribuciones son iguales, su divergencia KL será cero.
No simetría: La divergencia KL no es simétrica. Esto quiere decir que D_KL(P || Q) no es igual a D_KL(Q || P). En otras palabras, la información que se pierde al aproximar P con Q no es la misma que la que se pierde al aproximar Q con P.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist
from scipy.stats import norm, entropy
from scipy.spatial.distance import jensenshannon

# Calcular la matriz de distancia euclidiana con un parámetro random_state
euclidean_distances = pdist(X_scaled.toarray(), metric='euclidean')
# Ajustar una distribución normal a las distancias euclidianas
mu, std = norm.fit(euclidean_distances)

# Crear un conjunto de puntos equiespaciados que cubran todo el rango de distancias euclidianas
x = np.linspace(np.min(euclidean_distances), np.max(euclidean_distances), 100)

# Calcular las densidades de las distribuciones en los puntos equiespaciados
p = norm.pdf(x, mu, std)
q = np.histogram(euclidean_distances, bins=100, density=True)[0]

#Calcular la distancia de Jensen-Shannon
js_distance = jensenshannon(p, q)

# # Calcular la divergencia KL
# kl_divergence = entropy(pk=p, qk=q)

# Visualizar la distribución de las distancias euclidianas y la distribución normal ajustada
plt.figure(figsize=(10, 6))
plt.hist(euclidean_distances, bins=100, density=True, alpha=0.6, color='g', label="Distancias Euclidianas")
plt.plot(x, p, label="Distribución Normal")
plt.title(f"Distribución de Distancias Euclidianas y Distribución Normal (Distancia JS: {js_distance:.4f})")
plt.xlabel("Distancia Euclidiana")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

```python
#Imprimir las conclusiones
if js_distance < 0.1:

  print("las distancias euclidianas siguen una distribución normal. Esto significa que los textos tienen una variabilidad similar y no hay muchos outliers.")

else:

  print("Las distancias euclidianas no siguen una distribución normal. Esto significa que los textos tienen una variabilidad diferente y puede haber algunos outliers que se alejen mucho del resto.")
```

Si calculamos la divergencia **KL** entre la **distribución de tus datos** y una **distribución normal teórica**

Se obtiene una medida de cuánta información se pierde al aproximar tus datos con una distribución normal. 

Si la divergencia KL es baja, esto sugiere que la distribución normal se ajusta bien a tus datos.

si la divergencia KL es alta, significa que la distribución normal no es una buena aproximación para tus datos.

```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.stats import jarque_bera, chi2, skew, kurtosis, describe

# Convertir los datos dispersos en una matriz densa
dense_data = X_scaled.toarray()

#Obtener un resumen estadístico de los datos
stats = describe(dense_data)

print(f"Resumen estadístico de los datos:\n{stats}")

# Calcular la asimetría y la curtosis de los datos
skewness = skew(dense_data)

kurtosis = kurtosis(dense_data)

# Realizar la prueba de Jarque-Bera
jb_stat, jb_pvalue = jarque_bera(dense_data)

print(f"Estadístico de Jarque-Bera: {jb_stat:.4f}")

print(f"Valor p de Jarque-Bera: {jb_pvalue:.4f}")

# Confirmar con la prueba Chi-cuadrado
dof = 2 # Grados de libertad (2 para asimetría y curtosis)

alpha = 0.05 # Nivel de significancia

chi_critical = chi2.ppf(1 - alpha, dof)

print(f"Estadístico crítico de Chi-cuadrado: {chi_critical:.4f}")

#imprimir las conclusiones
if jb_stat < chi_critical:

  print("No hay evidencia suficiente para rechazar la hipótesis nula. Los datos parecen seguir una distribución normal.")

else:

  print("Se rechaza la hipótesis nula. Los datos no siguen una distribución normal.")

  print("Esto significa que se debe aplicar alguna transformación para normalizar los datos o usar otro algoritmo que no asuma esa distribución.")
```

#**Bloque 11: Puntuación de silhouette para k means**

Es importante tener en cuenta que el método del codo es una técnica heurística y puede no ser siempre preciso. En algunos casos, el "codo" puede no ser claramente visible o puede haber más de un "codo" en el gráfico. En tales casos, puede ser útil complementar el análisis con otras técnicas, como la **puntuación de Silhouette**, para seleccionar el número óptimo de grupos.

**Puntuación de Silhouette**

La puntuación de Silhouette es una métrica que evalúa la calidad del clustering de un conjunto de datos. Se utiliza para medir qué tan bien se han agrupado los puntos de datos y si el número de grupos elegido es apropiado. La puntuación de Silhouette varía entre -1 y 1, donde un valor más cercano a 1 indica un buen clustering y un valor cercano a -1 sugiere un mal clustering.

El coeficiente de silueta o puntuación de silueta es una métrica utilizada para calcular la bondad de una técnica de agrupación. Su valor oscila entre -1 y 1.

***1:**Significa que los grupos están bien separados entre sí y claramente diferenciados.

**0:** Significa que los conglomerados son indiferentes, o podemos decir que la distancia entre conglomerados no es significativa.

**-1:** significa que los clústeres se asignan de forma incorrecta.

La puntuación de Silhouette para un punto de datos se calcula como:

Silhouette = (b - a) / max(a, b)

Donde **"a"** es la cohesión y **"b"** es la separación.

**Cohesión (a):** Es la distancia promedio entre el punto de datos y todos los demás puntos en el mismo grupo. Una cohesión baja indica que el punto está cerca de otros puntos en su grupo.

**Separación (b):** Es la distancia promedio más baja entre el punto de datos y todos los puntos en cualquier otro grupo. Una separación alta indica que el punto está lejos de los puntos en otros grupos.

**Dividir por el máximo de 'a' y 'b**  asegura que el resultado de la puntuación de Silhouette esté en un rango normalizado, independientemente de la escala de las distancias en el conjunto de datos. Esto hace que la puntuación de Silhouette sea más interpretable y comparable entre diferentes conjuntos de datos o algoritmos de clustering.

Una puntuación de Silhouette cercana a 1 indica que el punto está bien agrupado y está lejos de otros grupos, lo que sugiere un buen clustering. Una puntuación cercana a 0 indica que el punto está cerca del límite entre dos grupos, lo que sugiere que podría ser asignado a otro grupo. Una puntuación negativa indica que el punto podría haber sido asignado incorrectamente a otro grupo, ya que está más cerca de puntos en otro grupo que de los puntos en su propio grupo.

La puntuación de Silhouette se puede calcular para todos los puntos de datos y luego promediar para obtener una medida global de la calidad del clustering. Esta puntuación promedio se utiliza a menudo para comparar diferentes algoritmos de clustering o para seleccionar el número óptimo de grupos en un algoritmo como K-means. **Un número de grupos que produce la puntuación de Silhouette más alta generalmente se considera la mejor opción para el conjunto de datos dado.**

La característica más emocionante del agrupamiento DBSCAN es que es resistente a los valores atípicos . Tampoco requiere que se indique el número de grupos de antemano, a diferencia de K-Means, donde tenemos que especificar el número de centroides.

**Conclusión**

Silhouette Score, como muchas otras métricas de evaluación de agrupamiento, es susceptible de error. Siempre que se utilice para citar el rendimiento del algoritmo, se debe estar seguro de que la métrica de distancia utilizada en el algoritmo es capaz de separar linealmente los datos.

**En los casos en que los conjuntos de datos no sean linealmente separables y las dimensiones de los conjuntos de datos sean muy altas, debemos tener cuidado al citar la distancia de Silhouette.**

Las técnicas de reducción de dimensionalidad se pueden utilizar para reducir los datos en dos dimensiones para su visualización. Como regla general, siempre que se utilicen algoritmos de agrupamiento basados ​​en la densidad, la distancia de la silueta puede no ser una métrica adecuada.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_words(cluster_centers, vectorizer, n_top_words=15, title=None):
    # Obtener las características (palabras) del vectorizador
    feature_names = vectorizer.get_feature_names_out()
    
    # Configurar el estilo de la gráfica
    sns.set(style="whitegrid")
    
    # Crear una figura y ejes para múltiples subgráficos
    fig, axes = plt.subplots(nrows=cluster_centers.shape[0], ncols=1, figsize=(10, 5 * cluster_centers.shape[0]))
    fig.tight_layout(pad=5.0)

    # Graficar las palabras principales para cada clúster
    for i, ax in enumerate(axes):
        # Obtener los índices de las palabras principales en el clúster
        top_word_indices = cluster_centers[i].argsort()[-n_top_words:][::-1]

        # Obtener las palabras y sus importancias
        top_words = [feature_names[j] for j in top_word_indices]
        top_words_importances = cluster_centers[i, top_word_indices]

        # Crear un gráfico de barras con las palabras principales y sus importancias
        sns.barplot(x=top_words_importances, y=top_words, palette='viridis', ax=ax)
        
        # Configurar el título del subgráfico
        ax.set_title(f'Cluster {i + 1}', fontsize=16)
        ax.set_xlabel('Importance', fontsize=14)
        ax.set_ylabel('Top words', fontsize=14)

    # Configurar el título principal de la figura
    if title:
        fig.suptitle(title, fontsize=20, y=1.05)

    # Mostrar el gráfico
    plt.show()

# Graficar las 15 palabras principales de cada clúster
plot_top_words(kmeans.cluster_centers_, tfidf_vectorizer, title='Top 15 words for each K-means cluster')
```

#Conclusiones

1. Supuestos implícitos: K-means, al igual que muchos algoritmos de aprendizaje automático, tiene supuestos implícitos sobre la estructura de los datos. En el caso de K-means, se asume que los grupos tienen **formas esféricas y varianzas similares.** La distribución normal cumple con estos supuestos, ya que es simétrica, con una forma de campana y la varianza es igual en todas las direcciones. Cuando los datos siguen una distribución normal, los supuestos implícitos de K-means se mantienen, y el algoritmo puede encontrar grupos de manera eficiente.

2. **Minimización de la varianza intra-cluster**: El objetivo de K-means es minimizar la suma de las distancias cuadradas entre los puntos y sus centroides dentro de cada grupo. En términos estadísticos, esto equivale a minimizar la varianza intra-cluster. Cuando los datos siguen una distribución normal, la mayoría de los puntos están cerca de la media, y minimizar la varianza es una estrategia efectiva para separar los datos en grupos distintos. En otras palabras, los grupos que K-means encuentra son aquellos que tienen las menores variaciones en sus distancias al centroide, lo que coincide con la característica de la distribución normal de que los datos están concentrados alrededor de la media.

3. **La estructura de los datos**: La distribución normal tiene una estructura específica en la que la densidad de los datos disminuye a medida que nos alejamos de la media. Esto significa que hay regiones con menor densidad de datos entre los grupos, lo que facilita que K-means encuentre límites claros entre ellos. En comparación, si los datos tienen distribuciones no normales con formas irregulares o densidades variables, los límites entre los grupos pueden ser menos claros y K-means podría tener dificultades para encontrar una partición adecuada.

**A pesar de que K-means funciona bien con distribuciones normales**, es importante considerar que no es la solución óptima para todos los problemas de clustering. **Cuando los datos tienen distribuciones no normales, formas irregulares, o diferentes densidades y tamaños de grupos, otros algoritmos de clustering**, como **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** o **clustering jerárquico**, pueden ser más apropiados, ya que no hacen supuestos tan estrictos sobre la estructura de los datos y pueden adaptarse a diferentes características de los grupos.
